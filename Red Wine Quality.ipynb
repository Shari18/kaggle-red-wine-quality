## Modelling

### Test and Train Split

```python
# Putting feature variable to X
X = df.drop('quality', axis=1)
y = df.quality
```

```python
# train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

### Feature Scaling

```python
# columns 
columns = X_train.columns
# define standard scaler
scaler = StandardScaler()
# fit and transform the numerical columns
X_train = scaler.fit_transform(X_train)
# converting into dataframe
X_train = pd.DataFrame(X_train)
# adding the columns back
X_train.columns = columns
X_train.head()
```

## Model Building

```python
# buidling the model using SVM using linear function
classifier = SVC(kernel = 'linear', random_state = 100)
classifier.fit(X_train,y_train)
```

```python
# predicting the test result
y_test_pred = classifier.predict(X_test)
```

```python
# evaluating the model using confusion metrics
metrics.confusion_matrix(y_true=y_test, y_pred=y_test_pred)
```

```python
# accuracy
print("accuracy", metrics.accuracy_score(y_test, y_test_pred))

# precision
print("precision", metrics.precision_score(y_test, y_test_pred))

# recall/sensitivity
print("recall", metrics.recall_score(y_test, y_test_pred))

#f1-score
print("f1-score", metrics.f1_score(y_test, y_test_pred))
```

As we can clearly see, SVC model always predicts the quality of wine as 'bad'. So, `precision` is 0 and accuracy is 0.86.<br> So, we need to increase the precision. So, basically accuracy is not correct metric to measure the model efficiency. As model predicts all wine quality as bad.<br>
This means there is huge data imbalance and we have to perform undersampling of majority class or oversampling minority class

## Resampling Techniques - Oversample minority class

```python
from sklearn.utils import resample

# Putting feature variable to X
X = df.drop('quality', axis=1)
y = df.quality

# setting up testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)

# concatenate our training sets back together
X = pd.concat([X_train,y_train], axis=1)

# separate minority and majority class
not_good_qualtity = X[X.quality == 0]
good_qualtity = X[X.quality == 1]

# upsample minority
quality_upsampled = resample(good_qualtity,
                            replace=True,# sample with replacement
                            n_samples=len(not_good_qualtity),# match number in majority class
                            random_state=100)

# combine majority and upsampled minority
upsampled = pd.concat([not_good_qualtity, quality_upsampled])

# check new class counts
upsampled.quality.value_counts()
```

After resampling, we have an equal ratio of data points for each class

```python
# trying linear SVM
y_train = upsampled.quality
X_train = upsampled.drop('quality', axis = 1)
classifier_2 = SVC(kernel = 'linear', random_state = 100)
classifier_2.fit(X_train,y_train)

upsampled_pred = classifier_2.predict(X_test)

# accuracy
print("accuracy", metrics.accuracy_score(y_test, upsampled_pred))

# precision
print("precision", metrics.precision_score(y_test, upsampled_pred))

# recall/sensitivity
print("recall", metrics.recall_score(y_test, upsampled_pred))

#f1-score
print("f1-score", metrics.f1_score(y_test, upsampled_pred))
```

Now the `precision` is increased to `0.38` from `0`. Also, `f1-score` is at 0.54 which is better than our previous model which is at `0`.<br>
Now, let us perform undersampling to check if it performs better here.

## Resampling Techniques - Undersampling majority class

```python
# undersampling majority class
quality_downsampled = resample(not_good_qualtity,
                              replace=False,# sample without replacement
                            n_samples=len(good_qualtity),# match number in minority class
                            random_state=100)

# combine minority and downsampled majority
downsampled = pd.concat([quality_downsampled,good_qualtity])

# check new class counts
downsampled.quality.value_counts()
```

After undersampling, we have an equal ratio of data points for each class.<br>
Let us perform the SVM and see how the classifier performs

```python
# trying linear SVM
y_train = downsampled.quality
X_train = downsampled.drop('quality', axis = 1)
classifier_3 = SVC(kernel = 'linear', random_state = 100)
classifier_3.fit(X_train,y_train)

downsampled_pred = classifier_3.predict(X_test)

# accuracy
print("accuracy", metrics.accuracy_score(y_test, downsampled_pred))

# precision
print("precision", metrics.precision_score(y_test, downsampled_pred))

# recall/sensitivity
print("recall", metrics.recall_score(y_test, downsampled_pred))

#f1-score
print("f1-score", metrics.f1_score(y_test, downsampled_pred))
```

As we can see, the `f1-score` is reduced from `0.54` to `0.53` from the previous upsampled model.<br>
Also, we have very less quantity of data to train the model.<br>
Now, let us try one more sampling which is similar to upsampling which is to create synthetic samples

## Generate Synthetic Samples

```python
from imblearn.over_sampling import SMOTE

# Putting feature variable to X
X = df.drop('quality', axis=1)
y = df.quality

# setting up testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)

sm = SMOTE(random_state=100)
X_train, y_train = sm.fit_sample(X_train, y_train)
```

We have genrated the synthetic samples, let us fit the model and see how it performs

```python
# trying linear SVM
smote_classifier = SVC(kernel = 'linear', random_state = 100)
smote_classifier.fit(X_train,y_train)

smote_pred = smote_classifier.predict(X_test)

# accuracy
print("accuracy", metrics.accuracy_score(y_test, smote_pred))

# precision
print("precision", metrics.precision_score(y_test, smote_pred))

# recall/sensitivity
print("recall", metrics.recall_score(y_test, smote_pred))

#f1-score
print("f1-score", metrics.f1_score(y_test, smote_pred))
```

Our `f1-score` has slightly increased compare to the downsampled model. <br>
Bur `precision` and `recall` remain the same<br>
From all of the above techniques, we can choose upsampling works best by comparing various `f1-score` scores.<br>
Let us take the upsamlpling model and tune the hyper parameters

## Tuning hyper parameters

```python
# putting features in X and y variable
y_train = upsampled.quality
X_train = upsampled.drop('quality', axis = 1)
# applying GridSearchCV to tune the hyperparameters
parameters = [{'C' : [1.1,1.2,1.5,1.7,1.8,1.9], 'kernel' : ['linear']},
              {'C' : [1.1,1.2,1.5,1.7,1.8,1.9], 'kernel' : ['rbf'], 'gamma':[1.0,1.1,1.2,1.3,1.4,1.5,1.6]}]
grid_search = GridSearchCV(estimator = SVC(), param_grid = parameters, scoring = 'accuracy', cv = 10, n_jobs = -1)
grid_search = grid_search.fit(X_train, y_train)

# fit the model 
grid_search.fit(X_train, y_train)
```

```python
# displaying the result of grid search cv
pd.set_option('display.max_columns', 500)
cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results
```

```python
# print the best score and best parameter values
print("Best Score {0}".format(grid_search.best_score_))
print("Best Parameters {0}".format(grid_search.best_params_))
```

## Model Evaluation

```python
# specify optimal hyperparameters
best_params = {"C": 1.1, "gamma": 1.5, "kernel":"rbf"}

# model building
classifier_final = SVC(C=best_params['C'], gamma=best_params['gamma'], kernel=best_params['kernel'])
classifier_final.fit(X_train, y_train)
# predicting the model
y_pred_final = classifier_final.predict(X_test)

# metrics
print(metrics.confusion_matrix(y_test, y_pred_final), "\n")
print("accuracy", metrics.accuracy_score(y_test, y_pred_final))
print("precision", metrics.precision_score(y_test, y_pred_final))
print("sensitivity/recall", metrics.recall_score(y_test, y_pred_final))
#f1-score
print("f1-score", metrics.f1_score(y_test, smote_pred))
```

Now the `precision` is increased to `75%` from `0%`. It is a good improvement compared to linear SVC model.
